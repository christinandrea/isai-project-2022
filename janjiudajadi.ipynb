```python
import os
import pandas as pd
import cv2
from google.colab.patches import cv2_imshow
from tkinter import Tcl
import numpy as np
```

```python
df = pd.DataFrame()

for i in range (1,26):
  df['ndvi'+str(i)] = []

for i in range (1,26):
  df['ngrdi'+str(i)] = []
  
for i in range (1,26):
  df['gndvi'+str(i)] = []

for i in range (1,26):
  df['ci'+str(i)] = []
```

```python
#function 1

def func1(citra):
  dim = (5, 5)
  resized = cv2.resize(citra, dim, interpolation = cv2.INTER_AREA)

  red = resized[:,:,2]
  green = resized[:,:,1]
  nir = resized[:,:,0]

  #hij = rgn
  ndvi = ((nir.astype(float)-red)/(nir.astype(float)+red))
  ngrdi = (green.astype(float)-red)/(green.astype(float)+red)
  gndvi = (nir.astype(float)-green)/(nir.astype(float)+green)
  ci = (red-green.astype(float))/(red+green.astype(float))

  #dijadikan 1 dimensi
  ndvi = ndvi.flatten()
  ngrdi = ngrdi.flatten()
  gndvi = gndvi.flatten()
  ci = ci.flatten()

  record = []

  for i in ndvi:
    record.append(i)
  for i in ngrdi:
    record.append(i)
  for i in gndvi:
    record.append(i)
  for i in ci:
    record.append(i)

  df.loc[len(df.index)] = record

  
```

```python
def resamplingmean(channel,ls):
  x = 2.5
  y = 2.5
  count = 0
  for i in range (25):
    count += 1
    imgx = channel

    # print(x, y)
    img0 = cv2.getRectSubPix(imgx, (5,5), (x ,y))

    if(x == 22.5):
      x = 2.5
      y += 5.0
    else:
      x += 5.0

    ls.append(int(img0.mean()))
```

```python
#function 2

def func2 (citra):
  size0 = citra.shape[0] / 2
  size1 = citra.shape[1] / 2
  img0 = cv2.getRectSubPix(citra, (25,25), (size1 ,size0))
  nir = []
  red = []
  green = []
  n = img0[:,:,0]
  r = img0[:,:,2]
  g = img0[:,:,1]
  resamplingmean(r,red)
  resamplingmean(n,nir)
  resamplingmean(g,green)
  red = np.array(red)
  nir = np.array(nir)
  green = np.array(green)

  #hij = rgn
  ndvi = ((nir.astype(float)-red)/(nir.astype(float)+red))
  ngrdi = (green.astype(float)-red)/(green.astype(float)+red)
  gndvi = (nir.astype(float)-green)/(nir.astype(float)+green)
  ci = (red-green.astype(float))/(red+green.astype(float))

  #dijadikan 1 dimensi
  ndvi = ndvi.flatten()
  ngrdi = ngrdi.flatten()
  gndvi = gndvi.flatten()
  ci = ci.flatten()

  record = []

  for i in ndvi:
    record.append(i)
  for i in ngrdi:
    record.append(i)
  for i in gndvi:
    record.append(i)
  for i in ci:
    record.append(i)

  df.loc[len(df.index)] = record
```

```python
def resamplingmedian(channel,ls):
  x = 2.5
  y = 2.5
  count = 0
  for i in range (25):
    count += 1
    imgx = channel

    # print(x, y)
    img0 = cv2.getRectSubPix(imgx, (5,5), (x ,y))

    if(x == 22.5):
      x = 2.5
      y += 5.0
    else:
      x += 5.0

    ls.append(int(np.median(img0)))
```

```python
#function 3

def func3 (citra):
  size0 = citra.shape[0] / 2
  size1 = citra.shape[1] / 2
  img0 = cv2.getRectSubPix(citra, (25,25), (size1 ,size0))
  nir = []
  red = []
  green = []
  n = img0[:,:,0]
  r = img0[:,:,2]
  g = img0[:,:,1]
  resamplingmedian(r,red)
  resamplingmedian(n,nir)
  resamplingmedian(g,green)
  red = np.array(red)
  nir = np.array(nir)
  green = np.array(green)

  #hij = rgn
  ndvi = ((nir.astype(float)-red)/(nir.astype(float)+red))
  ngrdi = (green.astype(float)-red)/(green.astype(float)+red)
  gndvi = (nir.astype(float)-green)/(nir.astype(float)+green)
  ci = (red-green.astype(float))/(red+green.astype(float))

  #dijadikan 1 dimensi
  ndvi = ndvi.flatten()
  ngrdi = ngrdi.flatten()
  gndvi = gndvi.flatten()
  ci = ci.flatten()

  record = []

  for i in ndvi:
    record.append(i)
  for i in ngrdi:
    record.append(i)
  for i in gndvi:
    record.append(i)
  for i in ci:
    record.append(i)

  df.loc[len(df.index)] = record
```

```python
def opencsv(directory,targetkolom):
  csvfiles = directory
  listcsv = os.listdir(csvfiles)
  listcsv = Tcl().call('lsort', '-dict', listcsv)

  C = []
  for i in listcsv:
    if i.endswith(".csv"):
      read = pd.read_csv(csvfiles+i)
      temp = read[targetkolom].values.tolist()
      C += temp

  return(C)
```

```python
root = "rootFolder/"
foldercsv = "csv/"
listFolder = os.listdir(root)
listFolder = Tcl().call('lsort', '-dict', listFolder)
functions = [func1,func2,func3]
# functions = [func1]

for folderCopy in listFolder:
  if folderCopy != ".ipynb_checkpoints":
    print(folderCopy)
    folderx = Tcl().call('lsort', '-dict', os.listdir(root+folderCopy))
    # print(folderx)
    for i in range(len(functions)):
        for folderHST in folderx:
          if folderHST != ".ipynb_checkpoints":
            # print(folderHST)
            foldery = Tcl().call('lsort', '-dict', os.listdir(root+folderCopy+"/"+folderHST))
            for citra in foldery:
              if citra != ".ipynb_checkpoints":
                if citra.endswith(".png") or citra.endswith(".jpg") or citra.endswith(".jpeg") or citra.endswith(".PNG"):
                  # print(citra)
                  # citra = cv2.imread("rootFolder/CP1/HST0_copy1/HST0_KL10_copy1.png")
                  citra = cv2.imread(root+folderCopy+"/"+folderHST+"/"+citra)
                  # print(citra)
                  functions[i](citra)

        C = opencsv(foldercsv,"C Total")
        # print(len(C))
        # print(df)
        df.insert(loc=0, column="C", value=C)
        print(len(df))
        df.to_csv("klaten_"+folderCopy+'fungsi%d'%i+'.csv', index=False)
        df = pd.DataFrame(columns=df.columns)
        df = df.drop('C', axis=1)
```
